Beta Distribution In Python

Different a and b values yield different distributional shapes.

```python
# Fixing a at 1 and increasing b by 1.


import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
import seaborn as sn

plt.figure(figsize=(3,3))
s = np.linspace(0, 1, 100)
prior1 = st.beta(1, 1)
prior2 = st.beta(1, 2)
prior3 = st.beta(1, 3)
prior4 = st.beta(1, 4)
prior5 = st.beta(1, 10)
plt.plot(prior1.pdf(s), label='Prior1')
plt.plot(prior2.pdf(s), label='Prior2')
plt.plot(prior3.pdf(s), label='Prior3')
plt.plot(prior4.pdf(s), label='Prior4')
plt.plot(prior5.pdf(s), label='Prior5')
plt.legend()
plt.show()
```

Drawing samples from normal distribution

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
A= np.random.normal(100, 15, 1000)
B= np.random.normal(90, 15, 1000)
plt.hist(A)
plt.hist(B)
```

Let's pick 100 random samples from each distribution and see how many times individual values from A are greater than the ones from B.

```python
Mean_A =np.zeros(1)
Mean_B =np.zeros(1)
Counter=[]
for i in range(100):
    Mean_A=np.append(Mean_A,random.choice(A))
    Mean_B=np.append(Mean_B,random.choice(B))
    if Mean_A[-1] > Mean_B[-1]:
        Counter.append(1)
print(len(Counter))
print(Mean_A.mean())
print(Mean_B.mean())
```

61 times out of 100 individual values turn out to be greater than the ones from B. If we run again, we get a different result.
